## My solution for problem 2

```python
# Your code starts here
c = 1  # the constant line y = 1

import matplotlib.pyplot as plt

# I prefer to show all views together so it's easy to read and grade.
fig, axes = plt.subplots(3, 1, figsize=(10, 12))

ax = axes[0]
ax.plot(x, y)
ax.axhline(c, color='red', linestyle='--', label='y = 1')
ax.set_title("Overview: f(x) and the line y = 1")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.grid(True)
ax.legend(loc='upper left')

# 2) Zoomed view – LEFT intersection

ax = axes[1]
ax.plot(x, y)
ax.axhline(c, color='red', linestyle='--', label='y = 1')
ax.set_xlim(-7, -4)   # adjust by eye if needed
ax.set_ylim(0, 5)     # keep the y=1 line clearly visible
ax.set_title("Zoomed view – Left intersection with y = 1")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.grid(True)

# 3) Zoomed view – RIGHT intersection

ax = axes[2]
ax.plot(x, y)
ax.axhline(c, color='red', linestyle='--', label='y = 1')
ax.set_xlim(2, 5)     # adjust by eye if needed
ax.set_ylim(0, 5)
ax.set_title("Zoomed view – Right intersection with y = 1")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.grid(True)

plt.tight_layout()
plt.show()
```
